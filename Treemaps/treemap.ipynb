# TREE MAP EXAMPLE

```python
import numpy as np
import matplotlib.pyplot as plt
import squarify

volume2020 =[1.1,1.2,1.7,3.1,3.3,3.9,4,4.1,4.2,4.3,4.4,5,5.1,5.6,6.1,11.4,15.6,15.7]
volume2017=[0.85,1.22,1.83,2.62,2.97,4.46,4.06,4.99,4.02,11.95,4.55,4.77,5.66,5.77,6.24,11.26,13.63,9.13]
alcaldias = ['MILPA ALTA',
'CUAJIMALPA',
'MAGDALENA CONTRERAS',
'TLAHUAC',
'XOCHIMILCO',
'MIGUEL HIDALGO',
'AZCAPOTZALCO',
'BENITO JUAREZ',
'IZTACALCO',
'OTROS ESTADOS',
'TLALPAN',
'VENUSTIANO CARRANZA',
'COYOACAN',
'CUAUHTEMOC',
'ALVARO OBREGON',
'GUSTAVO A. MADERO',
'IZTAPALAPA',
'ESTADO DE MEXICO']


labels1 = ["%s\n %s %% \n" % (label) for label  in zip(alcaldias, volume2020)] #creating the labels 
labels2 = ["%s\n %s %% \n" % (label) for label  in zip(alcaldias, volume2017)] 
```

```python
color_list = ['#1B78F2', '#23D9B7', '#D7F205',
              '#F2B705', '#d35158', '#8C43E6','#A7C8F2','#048ABF','#027368','#025951','#F25116'] #color list 
plt.rc('font', size=9)
fig = plt.figure(figsize=(27,18),dpi= 300)
plt.subplot(211)
plt.title("PERIODO 2020", fontsize=20)
squarify.plot(sizes=volume2020, label=labels1,
              color=color_list, alpha=0.5, bar_kwargs=dict(linewidth=3, edgecolor="#F2F2F2"))

plt.axis('off')

plt.subplot(212)
plt.title("PERIODO 2017-2019", fontsize=20)
squarify.plot(sizes=volume2017, label=labels2,
              color=color_list, alpha=0.5, bar_kwargs=dict(linewidth=3, edgecolor="#F2F2F2"))
plt.axis('off')
#fig.set_facecolor('#eeffee')

plt.show()
```
